# Importing TPOT and necessary libraries

```python
import pandas as pd 
from sklearn.model_selection import train_test_split
import torch
from tpot import TPOTClassifier
```

# Reading model data encoded and cleaned in obesity types eda notebook

```python
data = pd.read_csv("model data.csv")
```

```python
X = data.drop(columns="NObeyesdad")
y = data["NObeyesdad"]
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=7)
```

# Training TPOT classifier

```python
tpt  = TPOTClassifier(generations=10,verbosity=2)
tpt.fit(x_train,y_train)
```

```python
tpt.get_params()
```

```python
tpt.fitted_pipeline_
```

```python
tpt.score(x_test,y_test)
```
