# Importing h2o and initializing it

```python
import h2o
import os

from h2o.automl import H2OAutoML

h2o.init()
```

# Reading model data cleaned and encoded in obesity eda notebook

```python
cleaned_data_path = os.path.join("..", "data", "obesity_clean.csv")
data = h2o.import_file(cleaned_data_path)
```

```python
data
```

```python
x= data.columns[:-1]
y="NObeyesdad"
```

```python
aml = H2OAutoML(max_models=10,seed=7)
aml.train(x=x,y=y,training_frame=data) 
```
